```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc 

%matplotlib inline
```

```python
lgbm_pred = pd.read_csv('model_store/lgbm_model/lgbm_pred_0.csv')
dnn_pred = pd.read_csv('model_store/dnn_model/dnn_pred_0.csv')
```

```python
def plot_roc(labels, predict_prob, title):
    false_positive_rate,true_positive_rate,thresholds=roc_curve(labels, predict_prob)
    roc_auc=auc(false_positive_rate, true_positive_rate)
    plt.title(title + ' ROC')
    plt.plot(false_positive_rate, true_positive_rate,'b',label='AUC = %0.4f'% roc_auc)
    plt.legend(loc='lower right')
    plt.plot([0,1],[0,1],'r--')
    plt.ylabel('TPR')
    plt.xlabel('FPR')
```

```python
plt.figure(figsize=(10,8))
plot_roc(lgbm_pred.y_true, lgbm_pred.y_pred, 'lgbm')
```

```python
plt.figure(figsize=(10,8))
plot_roc(dnn_pred.y_true, dnn_pred.y_pred, 'DNN')
```
